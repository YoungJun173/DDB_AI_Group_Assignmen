```python
#Import Libraries --- IGNORE ---
import pandas as pd
import numpy as np
import pickle

from pathlib import Path
```

```python
#Reading the pickle file
External_Data_Directory = Path('/Users/johnpatrickbritten/Library/CloudStorage/OneDrive-HvA/HvA/Digital Driven Business Masters/AI Methods for Business/Assignment Files/')
file_name="Final_Movies_Dataset.pkl"

#Make sure the data directory exists
External_Data_Directory.mkdir(parents=True, exist_ok=True)

#Construct the full file path
file_path = External_Data_Directory / file_name

#Read the pickle file
try:
    if file_path.exists():
        with open(file_path, 'rb') as file:
            final_loaded_df = pickle.load(file)
        
        print(f"Dataframe successfully loaded from {file_path}")
        print(f"Loaded DataFrame shape: {final_loaded_df.shape}")
    
    else:
        print(f"File not found: {file_path}")
        print("Please check that the EDA notebook has been run and the path is correct.")

except Exception as e:
    print(f"An error occurred while loading the pickle file: {e}")
```
